```python
with open("minimapOutput_MT276325.sam","r") as IF:
    read=[line.split("\t") for line in IF if not line.startswith("@")]
```

```python
pdReads=pd.DataFrame(read)
```

```python
pdReads.columns=["QNAME","FLAG","RNAME","POS","MAPQ","CIGAR","RNEXT","PNEXT","TLEN","SEQ","QUAL","AUX1","AUX2","AUX3","AUX4","AUX5","AUX6","AUX7","AUX8","AUX9","AUX10","AUX11","AUX12","AUX13"]
pdMapped=pdReads[pdReads.RNAME=="MT276325.1"]
```

```python
def parseCigar(CIGAR,POS,QNAME):
    #print(POS)
    output=[]
    readString=""
    refPos=int(POS)
    qPos=int(POS)
    for i in CIGAR:
        if i=="M":
            refPos+=int(readString)
            qPos+=int(readString)
            output.append([QNAME,"M",int(readString),qPos,refPos])
            readString=""
        elif i=="D":
            refPos+=int(readString)
            qPos=qPos#Not found in query
            output.append([QNAME,"D",int(readString),qPos,refPos])
            readString=""
        elif i=="I":
            refPos=refPos #Not found in ref
            qPos+=int(readString)
            output.append([QNAME,"I",int(readString),qPos,refPos])
            readString=""
        elif i=="N":
            refPos+=int(readString) #Not counted
            qPos=qPos#Sequence Not found in Reference
            output.append([QNAME,"N",int(readString),qPos,refPos])
            readString=""
        elif i=="S":
            refPos=refPos
            qPos+=int(readString)
            output.append([QNAME,"S",int(readString),qPos,refPos])
            readString=""
        elif i=="H":
            refPos=refPos
            qPos+=int(readString)
            output.append([QNAME,"H",int(readString),qPos,refPos])
            readString=""
        else:
            readString="".join([readString,i])
    return(pd.DataFrame(output,columns=["qName","type","Length","qPos","refPos"]))
```
