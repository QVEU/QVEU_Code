```python
"""
Useful functions
"""

# Define TCID50 titer function
def calc_titer(temp_dataframe, base_dilution=0):
    titers = []
    for line_num in range(len(temp_dataframe)):
        #print(temp_dataframe.iloc[line_num])
        dils = [12] + [int(x) / 12 for x in temp_dataframe.iloc[line_num][3:]]
        effective_dil = 0
        #print(dils)
        for i in range(len(dils)):
            if dils[i] >= .5:
                if dils[i+1] <= .5:
                    prev_dil = dils[i]
                    next_dil = dils[i+1]
                    effect_dil = (prev_dil - 0.5) / (prev_dil - next_dil)
                    #print(i, prev_dil, next_dil, effect_dil)
                    titers.append(10 ** (effect_dil + i + base_dilution))
                    break
            else:
                #print(i)
                titers.append(0)#np.NaN)
                break
    return titers

# Define function for extracting consensus and >50% mutants
def Q20toconsensus(q20_df):
    """
    Read through dataframe of position nucleotide counts
    """
    
    pos2nuc = {0:'A', 1:'C', 2:'G', 3:'T'}
    # Make list of max nucleotides
    consensus_list = []
    sums = []
    for i in range(len(q20_df)):
        if sum(q20_df.iloc[i]) >= 20:
            new_nuc = pos2nuc[list(q20_df.iloc[i]).index(max(list(q20_df.iloc[i])))]
        else:
            new_nuc = 'N'
        sums.append(q20_df.iloc[i].sum())
    return consensus_list, sums

def unique_high_counts(strain, protein):
    unique_mut_df = passage_df[(passage_df['strain'] == strain) &
                            (passage_df['protein'] == protein) &
                            (passage_df['non-synonymous'] == 'non-synonymous') &
                            (passage_df['coverage'] > 20) &
                            (passage_df['passage'] == 6) &
                            (passage_df['nucleotide percent'] > 0.5)]['protein position']
    unique_muts = sorted(list(set(unique_mut_df)))
    
    display_df = passage_df[(passage_df['strain'] == strain) &
                            (passage_df['protein'] == protein) &
                            (passage_df['non-synonymous'] == 'non-synonymous') &
                            (passage_df['coverage'] > 20) &
                            (passage_df['passage'] == 6) &
                            (passage_df['nucleotide percent'] > 0.5)]
    unique_position_counts = []
    for temp in [33, 37]:
        for cell in ['A549', 'RD', 'SH-SY5Y']:
            new_obs = []
            temp_df = display_df[(display_df['cell'] == cell) &
                                 (display_df['temperature'] == temp)]
            for position in unique_muts:
                i = 0
                for rep in 'ABCDEFGH':
                    #print(temp_df[(temp_df['replicate'] == rep) & temp_df['protein position'] == position]['nucleotide percent'])
                    if len(temp_df[(temp_df['replicate'] == rep) & (temp_df['protein position'] == position)]) != 0:
                        i += 1
                new_obs.append(i)
            unique_position_counts.append(new_obs)
    
    unique_position_counts = np.array(unique_position_counts)
    large_unique_muts = []
    large_unique_mut_counts = []
    for i, mut in enumerate(unique_muts):
        if unique_position_counts.T[i].sum() > 3:
            large_unique_muts.append(int(mut))
            large_unique_mut_counts.append(list(unique_position_counts.T[i]))
    return np.array(large_unique_mut_counts).T, large_unique_muts

def mut_count(list_of_list_of_muts):
    temp_dict = {}
    for rep in list_of_list_of_muts:
        for mut in rep:
            if mut in temp_dict.keys():
                temp_dict[mut] += 1
            else:
                temp_dict[mut] = 1
    return temp_dict

def co_frequencies(df_1, df_2, mut_pair_1, mut_pair_2):
    co_frequency = []
    for passage in range(6):
        passage += 1
        #print(passage)
        #pass_df = passage_df[(passage_df['strain']=='Fermon') & (passage_df['passage'] == passage)]
        for cell in ['RD', 'A549', 'SH-SY5Y']:
            #cell_df = pass_df[pass_df['cell'] == cell]
            for temperature in [33, 37]:
                for replicate in 'ABCDEFGH':
                    mut1_df = df_1[(df_1['temperature'] == temperature) & 
                                   (df_1['cell'] == cell) &
                                   (df_1['passage'] == passage) &
                                   (df_1['protein'] == '3D') &
                                   (df_1['protein position'] == int(mut_pair_1[0])) &
                                   (df_1['mutant aa'] == mut_pair_1[1]) &
                                   (df_1['replicate'] == replicate)]['nucleotide percent'].to_numpy()
                    
                    mut2_df = df_2[(df_2['temperature'] == temperature) & 
                                   (df_2['cell'] == cell) &
                                   (df_2['passage'] == passage) &
                                   (df_2['protein'] == '3D') &
                                   (df_2['protein position'] == int(mut_pair_2[0])) &
                                   (df_2['mutant aa'] == mut_pair_2[1]) &
                                   (df_2['replicate'] == replicate)]['nucleotide percent'].to_numpy()
                    
                    if len(mut1_df) != 0 and len(mut2_df) != 0:
                        freq_1 = float(mut1_df[0])
                        freq_2 = float(mut2_df[0])
                    elif len(mut1_df) == 0 and len(mut2_df) != 0:
                        freq_1 = 0
                        freq_2 = float(mut2_df[0])
                    elif len(mut1_df) != 0 and len(mut2_df) == 0:
                        freq_1 = float(mut1_df[0])
                        freq_2 = 0
                    else:
                        freq_1 = 0
                        freq_2 = 0
                    co_frequency.append([freq_1, freq_2])
    return np.array(co_frequency)
```

```python
# Import DataFrame of all passage data
passage_df = pd.read_csv('../data/EV-D68_all_passage_data_v7.csv')
```

```python

fold_pval = []
for cell in ['RD', 'A549', 'SH-SY5Y']:
    for strain in ['US/MO/14-18947', 'Fermon']:
        for protein in ['VP4', 'VP2', 'VP3', 'VP1', '2A', '2B', '2C', '3A', '3B', '3C', '3D']:
            print(f'looking at {protein}')
            mean_fit = []

            protein_df = passage_df[(passage_df['strain']==strain) & 
                                    (passage_df['protein'] == protein) & 
                                    (passage_df['cell'] == cell)]
            max_val = protein_df['protein position'].drop_duplicates().max()

            for i in range(int(max_val)):
                position = i + 1
                pp_pos = int(protein_df[protein_df['protein position']==position]['polyprotein position'].drop_duplicates())
                WT_aa = protein_df[(protein_df['protein position'] == position)]['WT aa'].drop_duplicates().to_numpy()[0]
                muts = protein_df[(protein_df['coverage']>20) & 
                                  (protein_df['protein position'] == position)]['mutant aa'].drop_duplicates().to_numpy()

                position_df = protein_df[protein_df['protein position'] == position]
                for mut in muts:
                    if mut != WT_aa:
                        for temp in [33, 37]:
                            for rep in 'ABCDEFGH':
                                traj = position_df[(position_df['replicate'] == rep) & 
                                                   (position_df['temperature'] == temp) & 
                                                   (position_df['mutant aa'] == mut) &
                                                   (position_df['passage'] == 6)]['nucleotide percent'].dropna()
                                if len(traj) == 1:
                                    #print(traj)
                                    mean_fit.append([strain, protein, temp, WT_aa, position, pp_pos, mut, rep, float(traj.iloc[0])])
            fitness_df = pd.DataFrame(mean_fit)
            fitness_df.columns = ['strain', 'protein', 'temperature', 'WT', 'position', 'polyprotein position',
                                  'mutant', 'replicate', 'fitness']

            pvals = []
            fold = []
            for i in range(int(max_val)):
                position_df = fitness_df[fitness_df['position'] ==i+1]
                WT_aa = protein_df[(protein_df['protein position'] == position)]['WT aa'].drop_duplicates().to_numpy()[0]

                #print(position_df)
                for mut in position_df['mutant'].drop_duplicates():
                    mut_df = position_df[position_df['mutant']==mut]
                    fit_33 = mut_df[mut_df['temperature'] == 33]['fitness'].to_numpy()
                    fit_37 = mut_df[mut_df['temperature'] == 37]['fitness'].to_numpy()
                    #print(fit_33, fit_37)
                    #print(fit_33)

                    if fit_33.mean() != 0 and fit_37.mean() != 0:
                        new_pval = -np.log10(scipy.stats.f_oneway(fit_37, fit_33)[1])
                        new_fold_change = np.log2(fit_37.mean() / fit_33.mean())
                        fold_pval.append([cell, strain, protein, str(WT_aa), i+1, int(position_df['polyprotein position'].drop_duplicates()), mut, new_pval, new_fold_change])
                    pvals.append(new_pval)
                    fold.append(new_fold_change)
temp_P6_fold_df = pd.DataFrame(fold_pval)
temp_P6_fold_df.columns = ['cell', 'strain', 'protein', 'WT aa', 'position', 'polyprotein position', 'mutant', '-log10 pval', 'log2 fold-change']
```

```python
alternate_protein = ['VP4', 'VP3', '2A', '2C', '3B', '3D']
fermon_pp_pos = passage_df[(passage_df['protein'].isin(alternate_protein))][['protein', 'polyprotein position']]
fermon_pp_start = []
fermon_pp_stop = []
for protein in alternate_protein:
    start = fermon_pp_pos[fermon_pp_pos['protein'] == protein]['polyprotein position'].min()
    stop = fermon_pp_pos[fermon_pp_pos['protein'] == protein]['polyprotein position'].max()
    fermon_pp_start.append(start)
    fermon_pp_stop.append(stop)
    
alternate_protein = ['VP4', 'VP3', '2A', '2C', '3B', '3D']
mo_pp_pos = passage_df[(passage_df['protein'].isin(alternate_protein))][['protein', 'polyprotein position']]
mo_pp_start = []
mo_pp_stop = []
for protein in alternate_protein:
    start = mo_pp_pos[mo_pp_pos['protein'] == protein]['polyprotein position'].min()
    stop = mo_pp_pos[mo_pp_pos['protein'] == protein]['polyprotein position'].max()
    mo_pp_start.append(start)
    mo_pp_stop.append(stop)
```

```python
RdRp_df = passage_df[(passage_df['strain']=='Fermon') & (passage_df['protein']=='3D')]
correlation_stat = []
correlation_pval = []
for i, pair_1 in enumerate(RdRp_muts):
    print(i)
    temp_stat = []
    temp_pval = []
    for j, pair_2 in enumerate(RdRp_muts):
        if i < j:
            #freqs = co_frequencies(pair_1, pair_2)
            specific_mut_1 = RdRp_df[(RdRp_df['protein position']==int(pair_1[0])) & 
                                     (RdRp_df['mutant aa'] ==  pair_1[1])]#['nucleotide percent'].to_numpy()
            #print(specific_mut_1)
            specific_mut_2 = RdRp_df[(RdRp_df['protein position']==int(pair_2[0])) & 
                                     (RdRp_df['mutant aa'] ==  pair_2[1])]#['nucleotide percent'].to_numpy()
            freqs = co_frequencies(specific_mut_1, specific_mut_2, pair_1, pair_2)
            stat, pval = scipy.stats.pearsonr(freqs[:, 0], freqs[:, 1])
            temp_stat.append(stat)
            temp_pval.append(pval)
        else:
            temp_stat.append(0)
            temp_pval.append(0)
    correlation_stat.append(temp_stat)
    correlation_pval.append(temp_pval)
```
